```python
import tensorflow as tf
import numpy as np
import os
import zipfile
from PIL import Image
import matplotlib.pyplot as plt

# Download and extract the CelebA dataset
zip_file_path = tf.keras.utils.get_file('celeba-dataset.zip',
                                        'https://s3-us-west-1.amazonaws.com/udacity-dlnfd/datasets/celeba.zip',
                                        extract=True)
dataset_dir = os.path.join(os.path.dirname(zip_file_path), 'img_align_celeba')

# Function to load and preprocess images
def load_celeba_dataset(dataset_dir, image_size=(64, 64), num_images=None):
    image_files = os.listdir(dataset_dir)
    if num_images:
        image_files = image_files[:num_images]
    images = []
    for filename in image_files:
        filepath = os.path.join(dataset_dir, filename)
        image = Image.open(filepath)
        image = image.resize(image_size)
        image = np.array(image)
        images.append(image)
    return np.array(images)

# Load the dataset
celeba_images = load_celeba_dataset(dataset_dir, num_images=10000)  # Adjust num_images as needed

# Display sample images
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(celeba_images[i])
    plt.axis('off')
plt.show()
```
